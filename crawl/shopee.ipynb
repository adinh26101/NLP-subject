```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
class Shopee():
    def __init__(self, urls):
        self.urls = urls
        self.df = pd.DataFrame(columns = ['khach_cmt', 'rating', 'shop_phan_hoi'])
        self.driver = webdriver.Chrome(ChromeDriverManager().install())
        
    def row(self):
        return {'khach_cmt': '', 'rating': '', 'shop_phan_hoi': ''}

    def get_cmt(self, cmt):
        # customer cmt
        try:
            cus_cmt = cmt.find_element_by_class_name("_3NrdYc").text
        except:
            cus_cmt = ''
        # rating
        rating = len(cmt.find_elements_by_class_name("shopee-svg-icon.icon-rating-solid--active.icon-rating-solid"))
        # response
        try:
            shop_response = cmt.find_element_by_class_name("DW3H3c").text
        except:
            shop_response = ''
        return cus_cmt, rating, shop_response

    def crawl(self):
        for url in self.urls:
            self.driver.get(url)
            time.sleep(10)

            # tất cả# 5 sao# 4 sao# 3 sao# 2 sao# 1 sao# có bình luận# có ảnh/video
            options = self.driver.find_elements_by_class_name("product-rating-overview__filter")
            time.sleep(1)

            for i in range(3, 6):
                # pick sao 1-3 filter 
                options[i].click()
                time.sleep(4)
                
                page_past = ''
                still_have = True
                while(still_have):
                    try:
                        page_now = self.driver.find_element_by_class_name("shopee-button-solid.shopee-button-solid--primary").text
                    except:
                        break

                    if page_past == page_now:
                        break
                    else:
                        page_past = page_now

                    try:
                        # all cmt
                        all_cmt = self.driver.find_elements_by_class_name("shopee-product-rating__main")
                        
                        for cmt in all_cmt:
                            row = self.row()
                            row['khach_cmt'], row['rating'], row['shop_phan_hoi'] = self.get_cmt(cmt)
                            self.df = self.df.append(row, ignore_index=True)

                        self.driver.find_element_by_class_name("shopee-icon-button.shopee-icon-button--right").click()
                        time.sleep(4)
                    except:
                        self.driver.find_element_by_class_name("shopee-icon-button.shopee-icon-button--left").click()
                        time.sleep(1)
                        self.driver.find_element_by_class_name("shopee-icon-button.shopee-icon-button--right").click()
                        time.sleep(4)


    def data(self):
        return self.df
```
